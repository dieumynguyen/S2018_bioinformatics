I implemented this in Python using its hmmlearn module and I get something close to the manual calculation above: HHLLLLLLL. 

```python
from __future__ import division
import numpy as np
from hmmlearn import hmm
```

```python
states = ["H", "L"]
n_states = len(states)

observations = ["A", "C", "G", "T"]
n_observations = len(observations)

start_probability = np.array([0.5, 0.5])

transition_probability = np.array([
  [0.5, 0.5, 0.4],
  [0.5, 0.5, 0.6]
])

emission_probability = np.array([
    [0.2, 0.3, 0.3, 0.2],
    [0.3, 0.2, 0.2, 0.3],
])

model = hmm.MultinomialHMM(n_components=n_states)
model.startprob_=start_probability
model.transmat_=transition_probability
model.emissionprob_=emission_probability

# predict a sequence of hidden states based on visible states
sequence = np.array([[2,2,1,0,1,3,2,0,0]]).T

model = model.fit(sequence)
logprob, path = model.decode(sequence, algorithm="viterbi")
print("Observed sequence:", ", ".join(map(lambda x: observations[x], sequence.T[0])))
print("Probable path:", ", ".join(map(lambda x: states[x], path)))
```
